```python
import numpy as np
import matplotlib.pyplot as plt
```

## Draw vectors

```python
# initialize 2 vectors

color1 = [40,120,60]
color2 = [60,50,90]

ax = plt.figure().add_subplot(projection='3d')

# draw

# set the coordinates(red,green,blue)
ax.scatter(xs=color1[0], ys=color1[1], zs=color1[2], zdir='z', label='Color1')
ax.scatter(xs=color2[0], ys=color2[1], zs=color2[2], zdir='z', label='Color2')
# set origin [0,0,0]
ax.scatter(0,0,0, zdir='z', label='Origin', c='black')

# set arrow
ax.quiver(0,0,0, color1[0], color1[1], color1[2], color='black', arrow_length_ratio=0.1)
ax.quiver(0,0,0, color2[0], color2[1], color2[2], color='black', arrow_length_ratio=0.1)

# make legend, set axes limits and labels

ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# customize the view angle so it's easier to see that the scatter points lie on the plane y=0

ax.view_init(elev=20., azim=-35, roll=0)

plt.show
```

## How far these vectors are (distance between these vectors)

### Manhattan distance

```python
L1 = [color1[i] - color2[i] for i in range(len(color1))]
L1 = np.abs(L1).sum()
print('the color1 is at 120 units away from the color2', L1)

# alternate method
a = np.array(color1)
b = np.array(color2)
np.linalg.norm((a-b), ord=1)
```

### Euclidian distance ( the shortest distance between the 2 vectors point)

```python
L2 = [(color1[i] - color2[i])**2 for i in range(len(color1))]
L2 = np.sqrt(np.array(L2).sum())
print(L2)

# alternate method
np.linalg.norm((a-b), ord=2)
print(L2)
```

## Cosine distance

```python
cosine = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print(cosine)
```
